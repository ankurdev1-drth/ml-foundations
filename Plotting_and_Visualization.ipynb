**Annotations and Drawing on a Subplot

```python
# You can add annotations
# and text using the text, arrow, and annotate functions. text draws text at given
# coordinates (x, y) on the plot with optional custom styling
```

```python
#let’s plot the closing S&P 500 index price since 2007 ( YOU CAN DOWNLOAD THE spx.csv file from the officialy github of the writer of python for data analysis book)
```

```python
from datetime import datetime
fig, ax = plt.subplots()
data = pd.read_csv("spx.csv", index_col=0, parse_dates=True)
spx = data["SPX"]
spx.plot(ax=ax, color="black")
crisis_data = [
(datetime(2007, 10, 11), "Peak of bull market"),
(datetime(2008, 3, 12), "Bear Stearns Fails"),
(datetime(2008, 9, 15), "Lehman Bankruptcy")
]
for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75), #draw labels at the indicated x and y coordinates
    xytext=(date, spx.asof(date) + 225),
    arrowprops=dict(facecolor="black", headwidth=4, width=2,
    headlength=4),
    horizontalalignment="left", verticalalignment="top")
# Zoom in on 2007-2010
ax.set_xlim(["1/1/2007", "1/1/2011"])
ax.set_ylim([600, 1800])
ax.set_title("Important dates in the 2008-2009 financial crisis") # adds a main title to the plot.

#btw those who have watched the movie must have understood this clearly haha (THE BIG SHORT movie_name)
```

```python
# #matplotlib has objects that represent many
# common shapes, referred to as patches. full set is located in matplotlib.patches.
```

```python
# To add a shape to a plot, you create the patch object and add it to a subplot ax by
# passing the patch to ax.add_patch

fig, ax = plt.subplots()

rect = plt.Rectangle((0.2,0.75), 0.4,0.15, color = 'black',alpha = 0.3)
circ = plt.Circle((0.7,0.2), 0.15, color = 'blue', alpha = 0.3)
pgon = plt.Polygon([[0.15,0.15], [0.35, 0.4],[0.2, 0.6]], color = 'green', alpha = 0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

```python
#alpha is basically used for giving that transparency effect to the graph or shapes
```

**Facet Grids and Categorical Data

```python
#for datasets where we have additional grouping dimensions :
#one way to visualize data with many categorical variables is to use a facet grid
# seaborn catplot function will be used for this 
```

```python
sns.catplot(x="day", y="tip_pct", hue="time", col="smoker",
                kind="bar", data=tips[tips.tip_pct < 1])
```

```python
#we can  also expand the facet grid by adding one row per time value 
```

```python
sns.catplot(x="day", y="tip_pct", row="time", col="smoker", kind="bar", data=tips[tips.tip_pct <1])
```

```python
# #box plots (which show the median, quartiles, and outliers) 
```

```python
sns.catplot(x="tip_pct", y="day", kind="box", data=tips[tips.tip_pct < 0.5])
```
